```python
# Download from Kaggle (requires API key)
!kaggle datasets download -d mlg-ulb/creditcardfraud
!unzip -o creditcardfraud.zip

import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns

df = pd.read_csv('creditcard.csv')
print(f"Dataset shape: {df.shape}")
print(f"Fraud rate: {df['Class'].mean()*100:.4f}%")

# Check for missing values
print(f"Missing values: {df.isnull().sum().sum()}")

# Basic statistics
print(df['Amount'].describe())
```

```python
# Plot fraud distribution
plt.figure(figsize=(12, 5))

plt.subplot(1, 2, 1)
fraud_count = df['Class'].value_counts()
plt.bar(['Normal', 'Fraud'], fraud_count.values, color=['green', 'red'])
plt.title('Transaction Distribution')
plt.ylabel('Count')

plt.subplot(1, 2, 2)
fraud_amounts = df[df['Class'] == 1]['Amount']
normal_amounts = df[df['Class'] == 0]['Amount']
plt.hist(normal_amounts, bins=50, alpha=0.5, label='Normal', color='green')
plt.hist(fraud_amounts, bins=50, alpha=0.5, label='Fraud', color='red')
plt.xlabel('Transaction Amount')
plt.ylabel('Frequency')
plt.legend()
plt.title('Amount Distribution by Transaction Type')

plt.tight_layout()
plt.show()
```

```python
from sklearn.ensemble import IsolationForest
from sklearn.preprocessing import StandardScaler
from sklearn.model_selection import train_test_split
from sklearn.metrics import classification_report, confusion_matrix, roc_auc_score

# Prepare features
X = df.drop('Class', axis=1)
y = df['Class']

# Scale features
scaler = StandardScaler()
X_scaled = scaler.fit_transform(X)

# Train Isolation Forest
iso_forest = IsolationForest(
    contamination=0.001,  # Expected fraud rate
    random_state=42,
    n_estimators=100,
    max_samples='auto'
)

iso_forest.fit(X_scaled)

# Predict (-1 = fraud, 1 = normal)
df['anomaly_score'] = iso_forest.decision_function(X_scaled)
df['fraud_pred'] = (iso_forest.predict(X_scaled) == -1).astype(int)

print("="*60)
print("ISOLATION FOREST RESULTS")
print("="*60)
print(classification_report(y, df['fraud_pred']))

# Confusion Matrix
cm = confusion_matrix(y, df['fraud_pred'])
sns.heatmap(cm, annot=True, fmt='d', cmap='Blues')
plt.xlabel('Predicted')
plt.ylabel('Actual')
plt.title('Confusion Matrix - Isolation Forest')
plt.show()
```

```python
import tensorflow as tf
from tensorflow.keras.models import Model
from tensorflow.keras.layers import Input, Dense, Dropout
from sklearn.preprocessing import MinMaxScaler

# Normalize to [0,1] for autoencoder
minmax_scaler = MinMaxScaler()
X_scaled_mm = minmax_scaler.fit_transform(X)

# Split into train (normal only) and test
X_train = X_scaled_mm[y == 0]  # Train on normal transactions only
X_test = X_scaled_mm

# Autoencoder architecture
input_dim = X_scaled_mm.shape[1]

input_layer = Input(shape=(input_dim,))

# Encoder
encoded = Dense(14, activation='relu')(input_layer)
encoded = Dense(7, activation='relu')(encoded)
encoded = Dense(3, activation='relu')(encoded)

# Decoder
decoded = Dense(7, activation='relu')(encoded)
decoded = Dense(14, activation='relu')(decoded)
decoded = Dense(input_dim, activation='sigmoid')(decoded)

autoencoder = Model(input_layer, decoded)
autoencoder.compile(optimizer='adam', loss='mse')

# Train
history = autoencoder.fit(
    X_train, X_train,
    epochs=30,
    batch_size=256,
    validation_split=0.1,
    shuffle=True,
    verbose=1
)

# Plot training loss
plt.plot(history.history['loss'], label='Training Loss')
plt.plot(history.history['val_loss'], label='Validation Loss')
plt.xlabel('Epoch')
plt.ylabel('Loss')
plt.legend()
plt.title('Autoencoder Training')
plt.show()

# Calculate reconstruction error
reconstructions = autoencoder.predict(X_test)
mse = np.mean(np.square(X_test - reconstructions), axis=1)

# Set threshold at 99.5th percentile of normal transactions
threshold = np.percentile(mse[y == 0], 99.5)
df['reconstruction_error'] = mse
df['fraud_ae'] = (mse > threshold).astype(int)

print("="*60)
print("AUTOENCODER RESULTS")
print("="*60)
print(classification_report(y, df['fraud_ae']))
```

```python
# Ensemble: fraud if either model detects it
df['fraud_ensemble'] = (df['fraud_pred'] == 1) | (df['fraud_ae'] == 1)

print("="*60)
print("ENSEMBLE RESULTS")
print("="*60)
print(classification_report(y, df['fraud_ensemble']))

# Feature importance from Isolation Forest
feature_importance = pd.DataFrame({
    'feature': X.columns,
    'importance': iso_forest.feature_importances_
}).sort_values('importance', ascending=False)

print("\nTop 10 Important Features:")
print(feature_importance.head(10))
```

```python
%%writefile app.py
from fastapi import FastAPI, HTTPException
from pydantic import BaseModel
import joblib
import numpy as np
import pandas as pd
from tensorflow.keras.models import load_model

app = FastAPI(title="Fraud Detection API")

# Load models
iso_forest = joblib.load('isolation_forest.pkl')
scaler = joblib.load('scaler.pkl')
minmax_scaler = joblib.load('minmax_scaler.pkl')
autoencoder = load_model('autoencoder.h5')
threshold = joblib.load('threshold.pkl')

class Transaction(BaseModel):
    V1: float
    V2: float
    V3: float
    V4: float
    V5: float
    V6: float
    V7: float
    V8: float
    V9: float
    V10: float
    V11: float
    V12: float
    V13: float
    V14: float
    V15: float
    V16: float
    V17: float
    V18: float
    V19: float
    V20: float
    V21: float
    V22: float
    V23: float
    V24: float
    V25: float
    V26: float
    V27: float
    V28: float
    Amount: float

@app.get("/")
def root():
    return {"message": "Fraud Detection API", "status": "running"}

@app.post("/predict")
def predict(transaction: Transaction):
    # Convert to array
    features = np.array([[getattr(transaction, f'V{i}') for i in range(1, 29)] + [transaction.Amount]])

    # Scale
    features_scaled = scaler.transform(features)
    features_mm = minmax_scaler.transform(features)

    # Isolation Forest prediction
    iso_pred = iso_forest.predict(features_scaled)[0] == -1

    # Autoencoder prediction
    recon = autoencoder.predict(features_mm, verbose=0)
    mse = np.mean(np.square(features_mm - recon), axis=1)[0]
    ae_pred = mse > threshold

    # Ensemble
    is_fraud = bool(iso_pred or ae_pred)

    return {
        "is_fraud": is_fraud,
        "isolation_forest": bool(iso_pred),
        "autoencoder": bool(ae_pred),
        "reconstruction_error": float(mse),
        "risk_score": float(1 - iso_forest.decision_function(features_scaled)[0])
    }
```
